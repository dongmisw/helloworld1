```python
import pandas as pd
```

```python
df_score = pd.DataFrame(
    {
        '이름'    : ['홍길동','변우석','카리나','송혜교'],
        '영어점수' : [50,100,90,95],
        '수학점수' : [30,40,50,60]
    }
)
```

```python
df_score
```

```python
print(df_score.index)
print(df_score.columns)
print(df_score.values)
```

```python
df_score
```

```python
# 수학 점수 평균 
# built-in function  파이썬이 기본적으로 제공하는 function 
sum(df_score['수학점수'])/4
```

```python
# 수학 점수 평균
# pandas의 DataFrame에서 제공하는 함수 사용
df_score['수학점수'].sum() / df_score['수학점수'].count()
```

```python
df_score['수학점수'].mean()
```

```python
df_score['수학점수']
```

```python
df_score['수학점수']>50
```

```python
(df_score['수학점수']>50).sum()
```

```python
# 수학, 영어점수가 다 궁금함 - 담임선생님
(df_score[['영어점수', '수학점수']] >50 ).sum()
```
